# **SVM**

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split,GridSearchCV, StratifiedShuffleSplit
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X1, Y1, test_size=0.2, random_state=42)

# Stratified ShuffleSplit for balancing class distribution (optional)
sss = StratifiedShuffleSplit(n_splits=5, test_size=0.2, random_state=42)

# Define hyperparameter grid (consider a more focused range)
param_grid = {
    'kernel': ['linear','rbf'],
    'C': [1, 10],  # Adjust range based on your data
    'gamma': [0.1, 1]  # Adjust range for chosen kernels
}

# Create and train SVM model with GridSearchCV (consider parallelization)
model_cv = GridSearchCV(SVC(), param_grid, cv=sss, n_jobs=-1)  # Use all cores
model_cv.fit(X1,Y1)

# Best model parameters
best_model = model_cv.best_estimator_
best_params = model_cv.best_params_

print("Best Hyperparameters:", best_params)

# Make predictions on test set with best model
y_pred = best_model.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)

# Additional performance metrics
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1)

# Plot confusion matrix
from sklearn.metrics import confusion_matrix


cm = confusion_matrix(y_test, y_pred)

fig, ax = plt.subplots()
ax.imshow(cm, cmap='Blues')
ax.set_xticks(range(len(np.unique(y_test))))  # Assuming unique class labels
ax.set_yticks(range(len(np.unique(y_test))))
ax.set_xlabel('Predicted Label')
ax.set_ylabel('True Label')
ax.set_title('Confusion Matrix')
for i in range(len(np.unique(y_test))):
    for j in range(len(np.unique(y_test))):
        ax.text(j, i, cm[i, j], ha='center', va='center', color='black')
fig.show()


from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_curve, average_precision_score, matthews_corrcoef, fbeta_score, cohen_kappa_score

# Train the model with the best hyperparameters
svm_model = SVC(kernel='rbf', C=10, gamma=0.1)
svm_model.fit(X_train, y_train)

# Make predictions
y_pred = svm_model.predict(X_test)
y_prob = svm_model.decision_function(X_test)  # Probability estimates for ROC curve

from sklearn.preprocessing import label_binarize

# Binarize the labels for multi-class classification
y_test_bin = label_binarize(y_test, classes=np.unique(Y1))
n_classes = y_test_bin.shape[1]

for i in range(n_classes):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_prob[:, i])
    roc_auc = roc_auc_score(y_test_bin[:, i], y_prob[:, i])
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f}) for Class {i}')
    plt.plot([0, 1], [0, 1], 'r--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve for Class {i}')
    plt.legend()
    plt.show()

from sklearn.metrics import precision_recall_curve, average_precision_score

# Compute precision and recall for each class separately
precision = dict()
recall = dict()
average_precision = dict()
for i in range(n_classes):
    precision[i], recall[i], _ = precision_recall_curve(y_test_bin[:, i], y_prob[:, i])
    average_precision[i] = average_precision_score(y_test_bin[:, i], y_prob[:, i])

# Plot Precision-Recall curve and compute average precision for each class
plt.figure(figsize=(8, 6))
for i in range(n_classes):
    plt.plot(recall[i], precision[i], label=f'Class {i} (AP = {average_precision[i]:.2f})')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.show()


# Other performance metrics
mcc = matthews_corrcoef(y_test, y_pred)
from sklearn.metrics import fbeta_score

# Compute F-beta score for each class and then compute the average
f_beta = fbeta_score(y_test, y_pred, beta=1, average='weighted')

print(f"F-beta Score for rbf kernel: {f_beta}")

kappa = cohen_kappa_score(y_test, y_pred)
print(f"MCC for rbf kernel: {mcc}")
print(f"F-beta Score for rbf kernel: {f_beta}")
print(f"Cohen's Kappa for rbf kernel: {kappa}")
```
